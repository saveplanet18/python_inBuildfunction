```python
class Normal_Queue(object):
    def __init__(self,elements):
        self.elements = elements
        self.front = -1
        self.rear = -1
        self.Queue = [None] * elements

    def Insertion_at_Rear(self,ele):
        if (self.rear +1) == self.elements:
            print("Queue is Overflow")
        elif (self.front == -1 and self.rear == -1):
            self.front = 0
            self.rear = 0
        else:
            self.rear += 1
        temp = self.Queue[self.rear] = ele
        return temp
    
    def Traveser_Linked_Queue(self):
        if (self.front == -1 and self.rear == -1):
            print("Queue is Empty")
        else:
            while (self.front != self.rear):
                print(self.Queue[self.front])
                self.front = (self.front + 1) % self.elements
            print(self.Queue[self.rear])


if __name__ == "__main__":
    root = Normal_Queue(9)
    root.Insertion_at_Rear(45)
    root.Insertion_at_Rear(36)
    root.Insertion_at_Rear(18)
    root.Traveser_Linked_Queue()
```

```python
class Circular_Queue(object):
    def __init__(self,elements):
        self.elements = elements
        self.front = -1
        self.rear = -1
        self.Queue = [None] * elements

    def Insertion_at_Start(self,ele):
        if (self.rear + 1) % self.elements == self.front:
            print("Queue is overflow")
        elif (self.front == -1 and self.rear == -1):
            self.front = 0
            self.rear = 0
        else:
            self.rear = (self.rear + 1) % self.elements
        temp = self.Queue[self.rear] = ele
        return temp
    
    def Deletion_at_Front(self):
        if (self.rear + 1) % self.elements == self.front:
            print("Queue is Overflow")
        elif (self.front == self.rear):
            self.front = -1
            self.rear = -1
        else:
            temp = self.Queue[self.front]
            self.front += 1
            return temp
                 
    def Traverser_Linked_List(self):
        if (self.front == -1 and self.rear == -1):
            print("Queue is empty")
        else:
            while (self.front != self.rear):
                print(self.Queue[self.front])
                self.front = (self.front + 1) % self.elements
            print(self.Queue[self.rear])


if __name__ == "__main__":
    root = Circular_Queue(9)
    root.Insertion_at_Start(45)
    root.Insertion_at_Start(36)
    root.Insertion_at_Start(18)
    root.Deletion_at_Front()
    root.Deletion_at_Front()
    root.Traverser_Linked_List()
```

```python
class Insertion_DeQueue(object):
    def __init__(self,elements):
        self.elements = elements
        self.front = -1
        self.rear = -1
        self.Queue = [None] * elements

    def Insertion_at_Front(self,ele):
        if (self.front == 0 and self.rear == self.elements) or self.front == self.rear + 1:
            print("Queue is Overflow")
        elif (self.front == -1 and self.rear == -1):
            self.front = 0
            self.rear = 0
            temp = self.Queue[self.front] = ele
            return temp
        elif (self.front == 0):
            self.front = self.elements-1
            temp = self.Queue[self.front] = ele
            return temp
        else:
            self.front -= 1
            temp = self.Queue[self.front] = ele
            return temp
        
    def Insertion_at_End(self,ele):
        if (self.front == 0 and self.rear == self.elements) or self.front == self.rear + 1:
            print("Queue is Empty")
        elif (self.front == -1 and self.rear == -1):
            self.front = 0
            self.rear = 0
            temp = self.Queue[self.rear] = ele
            return temp
        elif (self.rear == self.elements-1):
            self.rear = 0
            temp = self.Queue[self.rear] = ele
            return temp
        else:
            self.rear += 1
            temp = self.Queue[self.rear] = ele
            return temp
        
    def Deletion_at_Front(self):
        if (self.front == 0 and self.rear == self.elements) or self.front == self.rear + 1:
            print("Queue is Overflow")
        elif (self.front == self.rear):
            self.front = -1
            self.rear = -1
            temp = self.Queue[self.front]
            return temp
        elif (self.front == self.elements-1):
            self.front = 0
            temp = self.Queue[self.front]
            return temp
        else:
            self.front = self.front+1
            temp = self.Queue[self.front]
            return temp
        
    def Deletion_at_rear(self):
        if (self.front == 0 and self.rear == self.elements) or self.front == self.rear + 1:
            print("Queue is Overflow")
        elif (self.front == self.rear):
            self.front = -1
            self.rear = -1
            temp = self.Queue[self.rear]
            return temp
        elif (self.rear == self.elements-1):
            self.rear = 0
            temp = self.Queue[self.rear]
            return temp
        else:
            self.rear -= 1
            temp = self.Queue[self.rear]
            return temp
     
    def Traversed_Queue(self):
        if (self.front == -1 and self.rear == -1):
            print("Queue is Empty")
        else:
            while (self.front != self.rear):
                print(self.Queue[self.front])
                self.front = (self.front + 1) % self.elements
            print(self.Queue[self.rear])

if __name__ == "__main__":
    root = Insertion_DeQueue(9)
    root.Insertion_at_Front(45)
    root.Insertion_at_Front(36)
    root.Insertion_at_Front(18)
    root.Insertion_at_End(89)
    root.Insertion_at_End(78)
    root.Insertion_at_End(90)
    root.Deletion_at_Front()
    root.Deletion_at_Front()
    root.Deletion_at_rear()
    root.Deletion_at_rear()
    root.Deletion_at_Front()
    root.Traversed_Queue()
```

```python
class Doubly_Creation_List(object):
    def __init__(self,elements):
        self.elements = elements
        self.Next = None
        self.Prev = None

class Doubly_Circular_List():
    def __init__(self):
        self.Head = None

    def Insertion_at_Start(self,ele):
        new_Object = Doubly_Creation_List(ele)
        if (self.Head == None):
            self.Head = new_Object
            new_Object.Next = self.Head
            return
        else:
            temp = self.Head
            while (temp.Next != self.Head):
                temp = temp.Next
            temp.Next = new_Object
            new_Object.Next = self.Head
            self.Head.Prev = new_Object
    def Insertion_at_End(self,ele):
        new_Object = Doubly_Creation_List(ele)
        if (self.Head == None):
            self.Head = new_Object
            new_Object.Next = self.Head
            self.Head.Prev = new_Object
            return
        else:
            temp = self.Head
            while (temp.Next != self.Head):
                temp = temp.Next
            temp.Next = new_Object
            new_Object.Next = self.Head

    def Deletion_at_Start(self):
        if (self.Head == None):
            self.Head = None
            return
        else:
            temp = self.Head
            while (temp.Next != self.Head):
                temp = temp.Next
            First_Node = temp.Next
            self.Head = self.Head.Next
            temp.Next = self.Head
            First_Node.Prev = None

    def Deletion_at_End(self):
        if (self.Head == None):
            First_Node = self.Head
            self.Head = self.Head.Next
            First_Node.Prev = None
            return
        else:
            temp = self.Head
            while (temp.Next.Next != self.Head):
                temp = temp.Next
            End_Node = temp.Next
            temp.Next = temp.Next.Next
            self.Head  = temp.Next
            End_Node.Prev = None

    def Traverse_Linked_List(self):
        temp = self.Head
        if (temp != None):
            print("elements are listed..")
            while (temp != None):
                print(temp.elements, end ='->')
                temp = temp.Next
                if (temp == self.Head):
                    break
            print()
        else:
            print("elements are empty")


if __name__ == "__main__":
    root = Doubly_Circular_List()
    root.Insertion_at_Start(45)
    root.Insertion_at_Start(36)
    root.Insertion_at_Start(18)
    root.Insertion_at_End(89)
    root.Insertion_at_End(68)
    root.Insertion_at_End(90)
    root.Deletion_at_Start()
    root.Deletion_at_Start()
    root.Deletion_at_End()
    root.Deletion_at_End()
    root.Traverse_Linked_List()
```

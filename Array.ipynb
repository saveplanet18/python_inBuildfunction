```python
def Insertion_at_start(arr , ele):
    n = len(arr)
    arr.append(None)
    for i in range(n-1, -1,-1):
        arr[i+1] = arr[i]
    arr[0] = ele
    return arr

if __name__ == "__main__":
    arr = [18,5,3,6]
    Insertion_at_start(arr,33)
    print(arr)
```

```python
def Insertion_at_End(arr,ele):
    n = len(arr)
    arr.append(None)
    for i in range(n-1,-1):
        arr[i-1] = arr[i]
    arr[-1] = ele
    return arr

if __name__ == "__main__":
    arr = [18,5,3,6]
    Insertion_at_End(arr,98)
    Insertion_at_End(arr,355)
    print(arr)
```

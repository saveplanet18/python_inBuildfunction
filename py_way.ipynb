```python
class Creation_Singly_Linked_List(object):
    def __init__(self,elements):
        self.elements = elements
        self.Next = None


class Singly_Linked_List():
    def __init__(self):
        self.Head = None

    def Insertion_at_Start(self,ele):
        ptr = Creation_Singly_Linked_List(ele)
        if (self.Head == None):
            self.Head = ptr
            return
        else:
            ptr.Next = self.Head
            self.Head = ptr
    
    def Insertion_at_specific_Position(self,pos,ele):
        ptr = Creation_Singly_Linked_List(ele)
        if (pos < 1):
            print("position should be greater than <= 1")
        elif (pos == 1 and self.Head == None):
            self.Head = ptr
            ptr.Next = self.Head
            self.Head = ptr
            return
        else:
            temp = self.Head
            for i in range(pos-1):
                if (temp.Next != None):
                    temp = temp.Next
            if (temp != None):
                ptr.Next = temp.Next
                temp.Next = ptr

    def Insertion_at_End(self,ele):
        ptr = Creation_Singly_Linked_List(ele)
        if (self.Head == None):
            self.Head = ptr
            return
        else:
            temp = self.Head
            while (temp.Next != None):
                temp = temp.Next
            temp.Next = ptr
    
    def Search_Node(self,item):
        temp = self.Head
        found = 0
        i = 0
        if (temp != None):
            while (temp != None):
                i += 1
                if (temp.elements == item):
                    found += 1
                    break
                temp = temp.Next
            if (found == 1):
                print(found,"elements are found ",i)
            else:
                print(found,"elements are not found",i)
    
    def Count_Node(self):
        temp = self.Head
        Count = 0
        while (temp.Next != None):
            Count += 1
            temp = temp.Next
        return Count

    def Deletion_at_Start(self):
        if (self.Head == None):
            self.Head = None
            return
        else:
            First = self.Head
            self.Head = self.Head.Next
            First = None
    
    def Deletion_at_specific_Position(self,pos):
        if (pos < 1):
            print("position is grater than <= 1")
        elif (pos == 1 and self.Head == None):
            First = self.Head 
            self.Head = self.Head.Next
            First = None
        else:
            temp = self.Head
            for i in range(pos-1):
                if (temp.Next != None):
                    temp = temp.Next
            if (temp != None):
                Position = temp.Next
                temp.Next = temp.Next.Next
                Position = None
    
    def Deletion_at_End(self):
        if (self.Head == None):
            First = self.Head
            self.Head = self.Head.Next
            First = None
            return
        else:
            temp = self.Head
            while (temp.Next.Next != None):
                temp = temp.Next
            End_Node = temp.Next
            temp.Next = temp.Next.Next
            End_Node = None

    def Traverse_Linked_List(self):
        temp = self.Head
        if (temp != None):
            print("elements are listed..")
            while (temp != None):
                print(temp.elements,end ='->')
                temp = temp.Next
            print()
        else:
            print("elements are empty")


if __name__ == "__main__":
    root = Singly_Linked_List()
    root.Insertion_at_Start(45)
    root.Insertion_at_Start(36)
    root.Insertion_at_Start(18)
    root.Insertion_at_specific_Position(1,89)
    root.Insertion_at_specific_Position(2,69)
    root.Insertion_at_specific_Position(3,82)
    root.Insertion_at_End(90)
    root.Insertion_at_End(35)
    root.Insertion_at_End(25)
    # root.Deletion_at_Start()
    # root.Deletion_at_Start()
    # root.Deletion_at_Start()
    # root.Deletion_at_specific_Position(2)
    # root.Deletion_at_specific_Position(1)
    # root.Deletion_at_specific_Position(3)
    # root.Deletion_at_End()
    # root.Deletion_at_End()
    # root.Deletion_at_End()
    root.Search_Node(69)
    print(root.Count_Node())
    root.Traverse_Linked_List()
```
